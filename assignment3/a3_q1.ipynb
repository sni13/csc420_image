#### Question 1.3 Laplacian of Gaussian

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
# img is a 2000 * 2000 image
# with a 100 * 100 black square in the center
img = np.ones((2000, 2000), dtype=np.uint8) * 255
img[950:1050, 950:1050] = 0
plt.imshow(img, cmap='gray')
```

```python
# Create Laplacian of Gaussian kernel
def getLoGKernel(kernel_size, sigma):
    grid = np.arange(-(kernel_size-1)//2, (kernel_size-1)//2+1)
    x, y = np.meshgrid(grid, grid)
    
    log = ((x**2 + y**2) / (2 * sigma**2))
    log = ((log - 1) * np.exp(-log)) / (np.pi * sigma**2)
    
    return log
```

```python
# Calculate magnitudes for sigma [1, 100]
magnitudes = []

for sigma in range(1, 101):
    kernel = getLoGKernel(2000, sigma)
    magnitude = np.sum(np.multiply(kernel, img))
    magnitudes.append(magnitude)

print(f'Sigma = {np.argmax(magnitudes)+1} maximizes the magnitude of the response')
plt.plot(range(1, 101), magnitudes)
plt.xlabel('Sigma')
plt.ylabel('Magnitude of Response')

plt.show()
```
