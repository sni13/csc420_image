```python
# import packages
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image

import cv2
```

### 1. Original Image & Canny Edge Detection

```python
# read image
img = cv2.imread('./toronto.jpg') # numpy array
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img2 = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# plot original image & canny edge detection output
fig = plt.figure(figsize=(15,15))
ax1 = fig.add_subplot(131)
ax2 = fig.add_subplot(132)
ax3 = fig.add_subplot(133)

ax1.imshow(img2)
ax1.title.set_text('Original Image'), ax1.set_xticks([]), ax1.set_yticks([])
edges = cv2.Canny(gray, threshold1=75, threshold2=100)
ax2.imshow(gray, cmap='gray')
ax2.title.set_text('Grayscale Image'), ax2.set_xticks([]), ax2.set_yticks([])
ax3.imshow(edges,cmap = 'gray')
ax3.title.set_text('Edge Image'), ax3.set_xticks([]), ax3.set_yticks([])

```

### 2. Add Gaussian Noise at Different Noise Variance Levels

```python
# Add Gaussian Noise to Images with different noise variance levels 
# set up "fig" to store subplots
fig = plt.figure(figsize=(20,16))
ax1 = fig.add_subplot(131)
ax2 = fig.add_subplot(132)
ax3 = fig.add_subplot(133)

# Low noise variance
mean = 0
var = 10
sigma = var ** 0.5
gaussian1 = np.random.normal(mean, sigma, (img.shape[0],img.shape[1], img.shape[2])) 
noisy_image1 = np.zeros(img.shape, np.float32)
noisy_image1 = img + gaussian1
ax1.imshow(noisy_image1)
ax1.title.set_text('Low Variance = 10'), ax1.set_xticks([]), ax1.set_yticks([])

# Med noise variance
mean = 0
var = 100
sigma = var ** 0.5
gaussian2 = np.random.normal(mean, sigma, (img.shape[0],img.shape[1], img.shape[2])) 
noisy_image2 = np.zeros(img.shape, np.float32)
noisy_image2 = img + gaussian2
ax2.imshow(noisy_image2)
ax2.title.set_text('Medium Variance = 100'), ax2.set_xticks([]), ax2.set_yticks([])

# High noise variance
mean = 0
var = 1000
sigma = var ** 0.5
gaussian3 = np.random.normal(mean, sigma, (img.shape[0],img.shape[1], img.shape[2])) 
noisy_image3 = np.zeros(img.shape, np.float32)
noisy_image3 = img + gaussian3
ax3.imshow(noisy_image3)
ax3.title.set_text('High Variance = 1000'), ax3.set_xticks([]), ax2.set_yticks([])
```

### 3. Canny Edge Detection on Image With Gaussian Noise

```python
# set up "fig" to store subplots
fig2 = plt.figure(figsize=(20,16))
ax1 = fig2.add_subplot(131)
ax2 = fig2.add_subplot(132)
ax3 = fig2.add_subplot(133)

# Low noise variance
noisy_cp1 = np.uint8(noisy_image1)
edges1 = cv2.Canny(noisy_cp1, threshold1=75, threshold2=100)
ax1.imshow(edges1, cmap = 'gray')
ax1.title.set_text('Low Variance = 10'), ax1.set_xticks([]), ax1.set_yticks([])

# Med noise variance
noisy_cp2 = np.uint8(noisy_image2)
edges2 = cv2.Canny(noisy_cp2, threshold1=75, threshold2=100)
ax2.imshow(edges2, cmap = 'gray')
ax2.title.set_text('Med Variance = 100'), ax2.set_xticks([]), ax2.set_yticks([])

# High noise variance
noisy_cp3 = np.uint8(noisy_image3)
edges3 = cv2.Canny(noisy_cp3, threshold1=75, threshold2=100)
ax3.imshow(edges3, cmap = 'gray')
ax3.title.set_text('High Variance = 1000'), ax3.set_xticks([]), ax3.set_yticks([])
```

### Observations:
- The higher the noise variance is, the worse Canny edge detection performs. 
- With low variance of 10, Canny can detect almost as good as in the original image; with med vairance of 100, the edges detected by Canny is harder to observe; with high variance of 1000, we can hardly see any edges.
- Variance of Gaussian determines the extend of smoothing. Increasing the noise variance of Gaussian reduces Canny detector's sensitivity to noise and loses finer details of images.
